```python
import os
from six.moves import urllib

DOWNLOAD_ROOT = "https://raw.githubusercontent.com/ard-data/2020-rki-impf-archive/master/"
IMPFUNG_PATH = "data/2_csv"
IMPFUNG_URL = DOWNLOAD_ROOT + IMPFUNG_PATH + "/region_HH.csv"

def fetch_impfung_data(impfung_url=IMPFUNG_URL, impfung_path=IMPFUNG_PATH):
    if not os.path.isdir(impfung_path):
        os.makedirs(impfung_path)
    target_path = os.path.join(impfung_path,"region_HH.csv")
    urllib.request.urlretrieve(impfung_url,target_path)
    
import pandas as pd
def load_impfung_data(impfung_path=IMPFUNG_PATH):
    csv_path = os.path.join(impfung_path, "region_HH.csv")
    return pd.read_csv(csv_path)    

fetch_impfung_data()
impfung = load_impfung_data()
impfung.head()
impfung.info()
impfung.describe()
```

```python
# Matplotlib
%matplotlib inline 
import matplotlib.pyplot as plt
impfung.hist(bins=5, figsize=(20,15))
plt.show()
```
